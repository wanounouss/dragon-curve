# The Dragon Curve

```python
dc.draw_curve(11)
```

This curve is the result of 11 folds of a strip of paper in the same direction. It results in a fractal structure called the dragon curve. Pretty neat!! 

The successive folds create corners on the paper strip in the up or down direction. In practice, it is quite hard to predict which corner direction will result at every fold, but fortunately, we can simulate the process and observe the results of folding.

We can even calculate how many corners are created after folding the paper strip 11 times:

```python
dc.nb_corners(11)
```

This is not quite 2023 corners... which is unfortunate... But maths is great! So we can calculate how many folds plus a bit would be needed to get exactly 2023 corners, whatever that means in practice :)

```python
dc.nb_folds(2023)
```

Almost 11!! So let's just say that 11 produces 2023 corners shall we? ;)

Now how many corners do 2023 folds produce?? 

```python
big_number = dc.nb_corners(2023)
print(big_number)
```

```python
len(str(big_number))
```

Yup... many!! In fact, this number has 609 digits! Also, assuming that the number of particles in the observable universe is about $3.3\times10^{80}$, our number of corners after 2023 folds is about $10^{526}$ times greater!! which puts things in perspectives... 

Also, do not attempt to draw a curve with 2023 folds!!! Python will crash! In fact, I do not recommend trying above 20 folds.
